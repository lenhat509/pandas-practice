```python
import pandas as pd
```

```python
stores = [{'watches': 2, 'bananas': 10, 'bike': 4}, {'eggs': 20, 'bananas': 2, 'bike': 10, 'milk': 20}]
```

```python
shop = pd.DataFrame(stores, index=['store1', 'store2'])
shop
```

```python
shop['milk'] = [10, 20]
shop
```

```python
shop['motobike'] = shop['bike'] + shop['watches']
shop
```

```python
shop['bananas'][['store1','store2']] = [11, 3]
shop
```

```python
shop.loc['store3'] = [1,2,3,4,5,6]
shop
```

```python
shop.loc['store4'] = {'watches': 0, 'eggs': 10, 'bike': 20}
shop
```

```python
store5 = {'watches': 10, 'bananas': 20, 'bike': 10}
```

```python
new_shop = pd.DataFrame(store5, index=['store5'])
new_shop
```

```python
shop = shop.append(new_shop)
shop
```

```python
shop['car'] = shop['motobike']
shop
```

```python
shop.loc['store6'] = shop.loc['store5'] / 2
shop
```

```python
shop.insert(2, 'mouse', [1,2,3,4,5,6])
```

```python
shop
```

```python
shop.pop('car')
shop
```

```python
shop.drop(['milk', 'motobike'], axis=1, inplace=True)
shop
```

```python
shop.drop('store5', axis=0, inplace=True)
shop
```

```python
shop.rename(index = {'store1': 'target'}, inplace=True)
shop
```

```python
shop.rename(columns={'watches': 'cereal', 'bike': 'car'}, inplace=True)
shop
```
